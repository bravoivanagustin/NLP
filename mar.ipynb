```python
import pandas as pd
from datasets import load_dataset
import numpy as np
```

```python
dataset = load_dataset("daniel2588/sarcasm", split="train")
```

```python
df_original = dataset.to_pandas()
```

```python
df_nuevo = df_original.rename(columns={
    'comment': 'text',
    'label': 'is_sarcastic',
    'parent_comment': 'context'
})[['text', 'is_sarcastic', 'context']]
```

```python
df_nuevo.head()
```

Agrego las columnas que no estan en mi dataset, pero las necesitamos para integrar todos

```python
df_nuevo['source'] = "daniel2588/sarcasm"
df_nuevo['task'] = "context"
```

```python
df_nuevo['degree_of_sarcasm'] = pd.NA
df_nuevo['paraphrase'] = pd.NA
df_nuevo['type'] = pd.NA
```

Reordenamos para respetar lo que mando gonza

```python
df_final = df_nuevo[[
    'text',
    'is_sarcastic',
    'degree_of_sarcasm',
    'paraphrase',
    'context',
    'type',
    'source',
    'task'
]]
```

Exporto

```python
nombre_archivo = "sarcasm_daniel2588.csv"
```

```python
df_final.to_csv(nombre_archivo, index=False, na_rep='')
```
